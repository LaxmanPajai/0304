```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define dataset (AND logic)
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Target for AND gate
y = np.array([0, 0, 0, 1])

# Step 2: Add bias to inputs
X_bias = np.c_[np.ones(X.shape[0]), X]

# Step 3: Initialize weights
weights = np.zeros(X_bias.shape[1])
learning_rate = 0.1
epochs = 10

# Step 4: Perceptron Training
for epoch in range(epochs):
    for i in range(len(X_bias)):
        z = np.dot(X_bias[i], weights)
        y_pred = 1 if z >= 0 else 0
        error = y[i] - y_pred
        weights += learning_rate * error * X_bias[i]

print("Trained Weights:", weights)

# Step 5: Plot Decision Region
fig, ax = plt.subplots()
for i in range(len(X)):
    if y[i] == 0:
        ax.scatter(X[i][0], X[i][1], color='red', label='Class 0' if i==0 else "")
    else:
        ax.scatter(X[i][0], X[i][1], color='blue', label='Class 1' if i==3 else "")

# Decision boundary: w0 + w1*x + w2*y = 0 â†’ y = -(w0 + w1*x)/w2
x_vals = np.linspace(-0.2, 1.2, 100)
y_vals = -(weights[0] + weights[1]*x_vals) / weights[2]
ax.plot(x_vals, y_vals, 'k--', label='Decision Boundary')

plt.title("Perceptron Decision Region")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.legend()
plt.show()
```
