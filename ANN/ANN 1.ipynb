```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def softmax(x):
    e_x = np.exp(x - np.max(x))  # Numerical stability
    return e_x / e_x.sum(axis=0)

# Input range for plotting
x = np.linspace(-10, 10, 100)

# Create subplots
plt.figure(figsize=(12, 8))

# Sigmoid
plt.subplot(2, 3, 1)
plt.plot(x, sigmoid(x), label="Sigmoid", color="blue")
plt.title("Sigmoid")
plt.grid(True)

# Tanh
plt.subplot(2, 3, 2)
plt.plot(x, tanh(x), label="Tanh", color="orange")
plt.title("Tanh")
plt.grid(True)

# ReLU
plt.subplot(2, 3, 3)
plt.plot(x, relu(x), label="ReLU", color="green")
plt.title("ReLU")
plt.grid(True)

# Leaky ReLU
plt.subplot(2, 3, 4)
plt.plot(x, leaky_relu(x), label="Leaky ReLU", color="red")
plt.title("Leaky ReLU")
plt.grid(True)

# Softmax (on a smaller input range)
plt.subplot(2, 3, 5)
x_softmax = np.linspace(-2, 2, 10)
y_softmax = softmax(x_softmax)
plt.stem(x_softmax, y_softmax)  # Fixed: removed use_line_collection
plt.title("Softmax")
plt.grid(True)

# Layout
plt.tight_layout()
plt.suptitle("Activation Functions in Neural Networks", fontsize=16, y=1.02)
plt.show()
```
